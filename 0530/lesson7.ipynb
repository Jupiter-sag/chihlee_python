```python
#open("student.txt")引數值的呼叫
try:
    with open("student.txt","r",encoding="utf-8") as file:
        data=file.read()
        print(data)
except Exception as e:
    print("發生錯誤：",e)
```

```python
#引數名稱的呼叫,可以不依順序，只要依照名稱
try:
    with open(file="student.txt",encoding="utf-8") as file:
        data=file.read()
        print(data)
except Exception as e:
    print("發生錯誤：",e)
```

```python
#open("student.txt")引數值的呼叫
try:
    with open("student.txt",'r',-1,encoding='utf-8') as file:
        data=file.read()
        print(data)
except Exception as e:
    print("發生錯誤：",e)
```

```python
#混合的呼叫
#前面可以使用引數值的呼叫，後面可以使用引數名稱的呼叫
try:
    with open("student.txt",encoding='utf-8',mode="r") as file:
        data=file.read()
        print(data)
except Exception as e:
    print("發生錯誤：",e)
```
