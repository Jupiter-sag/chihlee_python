```python
import pandas as pd

dataFrame=pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame.dtypes
len(dataFrame)
dataFrame.info()
```

```python
dataFrame.columns=["統計年","區域別","年底人口數","土地面積","人口密度"]#改變欄位名稱
df=dataFrame.drop(index=0)
df1=df.dropna()
df2=df1.drop('統計年',axis=1)
df2
```

```python
df2["年底人口數"] = pd.to_numeric(df2["年底人口數"], errors="coerce")
df2
# 3. 這時候再用 dropna()，就能把原本是 '…' 的東沙、南沙群島（以及其他原本就是 NaN 的列）一次清乾淨！
df2 = df2.dropna(subset=["年底人口數"])
df2.info()
# # 4. 最後順利轉換為整數
df2["年底人口數"] = df2["年底人口數"].astype("int64")
df2["土地面積"] = df2["土地面積"].astype("float")
df2["人口密度"] = df2["人口密度"].astype("int64")
df2
```
