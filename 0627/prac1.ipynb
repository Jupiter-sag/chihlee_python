```python
import numpy as np
import matplotlib.pyplot as plt
import math

t=np.arange(0,4,0.2)
x=t*math.pi
y1=np.cos(x)
y2=np.cos(2*x)

fig=plt.figure()
ax1=fig.add_subplot(1,2,1)
ax1.plot(x,y1,linestyle="-",color="green")
ax2=fig.add_subplot(1,2,2)
ax2.plot(x,y2,linestyle="--",color="magenta")
plt.show()
```
